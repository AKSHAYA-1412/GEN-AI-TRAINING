```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from rank_bm25 import BM25Okapi
import numpy as np
```

```python
documents = [
    "Retrieval Augmented Generation improves LLM responses using external data.",
    "Chroma is a vector database used to store embeddings.",
    "BM25 is a keyword-based search algorithm used in information retrieval.",
    "Hybrid retrieval combines vector similarity and keyword matching.",
    "Transformers convert text into numerical embeddings for semantic understanding."
]
```

```python
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vector_db = Chroma.from_texts(documents, embedding)
```

```python
tokenized_docs = [doc.lower().split() for doc in documents]
bm25 = BM25Okapi(tokenized_docs)
```

```python
query = "What is hybrid retrieval?"
```

```python
vector_results = vector_db.similarity_search_with_score(query, k=3)

print("🔹 VECTOR SEARCH RESULTS:\n")

for doc, score in vector_results:
    print(f"Text: {doc.page_content}")
    print(f"Score (distance): {score}\n")
```

```python
tokenized_query = query.lower().split()
bm25_scores = bm25.get_scores(tokenized_query)

top_n = np.argsort(bm25_scores)[::-1][:3]

print("BM25 KEYWORD SEARCH RESULTS:\n")

for idx in top_n:
    print(f"Text: {documents[idx]}")
    print(f"Score: {bm25_scores[idx]}\n")
```

```python
# Convert vector distance → similarity
vector_scores = {}
for doc, score in vector_results:
    similarity = 1 / (1 + score)  # convert distance to similarity
    vector_scores[doc.page_content] = similarity

# Normalize BM25
bm25_scores_norm = bm25_scores / np.max(bm25_scores)

# Combine scores
hybrid_scores = {}

for i, doc in enumerate(documents):
    bm25_part = bm25_scores_norm[i]
    vector_part = vector_scores.get(doc, 0)

    hybrid_scores[doc] = 0.5 * bm25_part + 0.5 * vector_part

# Sort
sorted_hybrid = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)[:3]

print("HYBRID RETRIEVAL RESULTS:\n")

for text, score in sorted_hybrid:
    print(f"Text: {text}")
    print(f"Score: {score}\n")
```

```python
print("✅ FINAL ANSWERS COMPARISON\n")

# Vector answer
vector_answer = vector_results[0][0].page_content

# Hybrid answer
hybrid_answer = sorted_hybrid[0][0]

print("👉 Vector Search Answer:")
print(vector_answer)

print("\n👉 Hybrid Search Answer:")
print(hybrid_answer)
```
