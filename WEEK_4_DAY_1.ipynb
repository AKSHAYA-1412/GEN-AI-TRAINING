```python
import os
from getpass import getpass

from langchain.agents import create_agent
from langchain.tools import tool
from langchain_openai import ChatOpenAI
```

```python
os.environ["OPENAI_API_KEY"] = getpass("Enter OpenAI API Key: ")
```

```python
model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)
```

```python
@tool
def generate_test_cases(requirement: str) -> str:
    """Generate software test cases"""

    prompt = f"""
You are a QA expert.

Generate test cases for:

{requirement}

Include:
1. Positive cases
2. Negative cases
3. Edge cases

Format:
TC_ID | Scenario | Expected Result
"""
    response = model.invoke(prompt)
    return response.content

```

```python
agent = create_agent(
    model=model,
    tools=[generate_test_cases],
    system_prompt="You are an AI Test Case Generator Agent."
)

```

```python
query = "Merchant Payment"

result = agent.invoke({
    "messages": [
        {"role": "user", "content": query}
    ]
})

print("\nGenerated Test Cases:\n")
print(result["messages"][-1].content)
```
