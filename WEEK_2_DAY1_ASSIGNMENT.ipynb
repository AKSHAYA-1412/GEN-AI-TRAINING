```python
import chromadb
from sentence_transformers import SentenceTransformer
```

```python
documents = [
    "Artificial Intelligence is the simulation of human intelligence in machines.",
    "Machine Learning is a subset of AI that allows systems to learn from data.",
    "Transformers are deep learning models used in NLP tasks.",
    "RAG combines retrieval with generation for better answers.",
    "Blockchain is a decentralized ledger technology."
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')#load pretrained model that can convert sentence to embeddings
```

```python
client = chromadb.Client()#creates database connection
```

```python
collection = client.create_collection(name="my_documents")#table inside db
```

```python
for i, doc in enumerate(documents):
    embedding = model.encode(doc).tolist()   # Convert text → vector

    collection.add(
        documents=[doc],                     # Store original text
        embeddings=[embedding],              # Store vector
        ids=[str(i)]
    )
```

```python
query = input("Enter your query: ")
query_embedding = model.encode(query).tolist()
print("\nTop Results:\n")
results = collection.query(
    query_embeddings=[query_embedding],  # Search using vector
    n_results=2                          # Top 2 results
)
for doc in results['documents'][0]:
    print("-", doc)
```
